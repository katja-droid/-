```python
import pandas as pd

# ЗЧИТУЄМО ФАЙЛ
df = pd.read_csv('./shopping_trends_updated.csv', encoding='latin-1')

df
```

```python
# Побудуйте гістограму віку клієнтів, використовуючи Seaborn.
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=False, color='skyblue')

plt.title('Гістограма віку клієнтів')
plt.xlabel('Вік')
plt.ylabel('Частота')

plt.show()
```

```python
# Створіть стовпчасту діаграму або графік розподілу, щоб порівняти, які 
# товари покупають клієнти найчастіше
#ДИВИМОСЬ, ЯКА БУВАЄ ЧАСТОТА
unique_values = df['Frequency of Purchases'].unique()

print(unique_values)
```

```python
grouped_data = df.groupby(['Item Purchased', 'Frequency of Purchases']).size().unstack(fill_value=0)

grouped_data.plot(kind='bar', stacked=True, figsize=(10, 6))

plt.title('Розподіл покупок за частотою і товаром')
plt.xlabel('Товари')
plt.ylabel('Кількість покупок')
plt.xticks(rotation=45)
plt.legend(title='Частота покупок')

plt.show()
```

```python
# Використовуйте scatter plot, щоб візуалізувати залежність між віком та 
# сумою покупок.
ages = df['Age']
purchase_amounts = df['Purchase Amount (USD)']

plt.figure(figsize=(8, 6))
plt.scatter(ages, purchase_amounts, color='blue', alpha=0.5)

plt.title('Залежність між віком та сумою покупок')
plt.xlabel('Вік')
plt.ylabel('Сума покупок')

plt.grid(True)

plt.show()
```

```python
# Побудуйте графік розподілу оцінок товарів для оцінки якості

ratings = df['Review Rating']

plt.figure(figsize=(8, 6))
plt.hist(ratings, bins=5, color='skyblue', edgecolor='black')

plt.title('Розподіл оцінок товарів')
plt.xlabel('Оцінка')
plt.ylabel('Кількість')

plt.show()
```

```python
# Створіть категоріальну діаграму, щоб відобразити розподіл покупок за 
# статтею.
gender = df['Gender']

plt.figure(figsize=(8, 6))
sns.countplot(x=gender, palette='muted')

plt.title('Розподіл покупок за статтю')
plt.xlabel('Стать')
plt.ylabel('Кількість')

plt.show()
```

```python
# Вивчіть, чи існує зв'язок між рейтингом товару і сумою покупок за 
# допомогою scatter plot.

ratings = df['Review Rating']
purchase_amounts = df['Purchase Amount (USD)']

plt.figure(figsize=(8, 6))
plt.scatter(ratings, purchase_amounts, color='blue', alpha=0.5)

plt.title('Зв\'язок між рейтингом товару і сумою покупок')
plt.xlabel('Рейтинг товару')
plt.ylabel('Сума покупок (USD)')

plt.grid(True)

plt.show()
```

```python
# Побудуйте стовпчату діаграму, яка відображає, в який сезон клієнти 
# роблять найбільше покупок.
purchase_counts = df.groupby('Season').size()

plt.figure(figsize=(8, 6))
purchase_counts.plot(kind='bar', color='skyblue')

plt.title('Сезонні покупки')
plt.xlabel('Сезон')
plt.ylabel('Кількість покупок')

plt.show()
```

```python
# Створіть графік розподілу розмірів одягу, щоб визначити популярність 
# різних розмірів.
import numpy as np
clothing_sizes = df['Size'].unique()
size_counts = {size: np.sum(df['Size'] == size) for size in clothing_sizes}

plt.figure(figsize=(8, 6))
plt.bar(size_counts.keys(), size_counts.values(), color='skyblue')

plt.title('Розподіл розмірів одягу')
plt.xlabel('Розмір')
plt.ylabel('Кількість')

plt.show()
```

```python
# Порівняйте, скільки покупок робиться за різними методами оплати за 
# допомогою стовпчатої діаграми.
payment_counts = df['Payment Method'].value_counts()

plt.figure(figsize=(10, 6))
payment_counts.plot(kind='bar', color='skyblue')

plt.title('Кількість покупок за різними методами оплати')
plt.xlabel('Метод оплати')
plt.ylabel('Кількість покупок')

plt.show()
```

```python
# Створіть scatter plot для визначення, чи існує зв'язок між кількістю 
# попередніх покупок та сумою поточної покупки.
previous_purchases = df['Previous Purchases']
current_purchase_amount = df['Purchase Amount (USD)']

plt.figure(figsize=(8, 6))
plt.scatter(previous_purchases, current_purchase_amount, color='blue', alpha=0.5)

plt.title('Зв\'язок між кількістю попередніх покупок та сумою поточної покупки')
plt.xlabel('Кількість попередніх покупок')
plt.ylabel('Сума поточної покупки (USD)')

plt.grid(True)
plt.show()
```

```python
# Побудуйте графік поверхні (3D surface plot) для візуалізації залежності 
# між трьома числовими змінними з використанням бібліотеки Seaborn.
# Налаштуйте відображення графіка, включаючи назву графіку, підписи 
# осей і інші необхідні параметри для кращого розуміння візуалізації.
# Додайте кольорову легенду.
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

data = {'x': x.flatten(),
        'y': y.flatten(),
        'z': z.flatten()}
df = pd.DataFrame(data)

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_trisurf(df['x'], df['y'], df['z'], cmap='viridis')

ax.set_title('Графік поверхні z = sin(sqrt(x^2 + y^2))')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

mappable = ax.collections[0]
plt.colorbar(mappable, ax=ax, label='Z')

plt.show()
```

```python
# Побудуйте графік теплової поверхні для візуалізації залежності між 
# двома або більше числовими змінними на основі матриці даних, 
# використовуючи бібліотеку Seaborn.

data_matrix = np.random.rand(10, 10)
sns.heatmap(data_matrix, cmap='viridis')
plt.title('Heatmap Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()

```
